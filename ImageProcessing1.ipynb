```python
from skimage import io
from skimage import color
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

image = io.imread('Images\\catcop.JPG')
io.imshow(image)
```

```python
right_half = image[:, image.shape[1]//2:]
io.imshow(right_half)
```

```python
zeroed_image = image.copy()
zeroed_image[:,:,[1, 2]] = 0
io.imshow(zeroed_image)
```

```python
io.imshow(image)
```

```python
gray_im = np.average(image, axis=2)
io.imshow(gray_im, cmap='gray')
```

```python
avg = (image[:,:,0]//3+image[:,:,1]//3+image[:,:,2]//3)
gray_im = np.zeros_like(image)
gray_im[:,:,0] = avg
gray_im[:,:,1] = avg
gray_im[:,:,2] = avg
io.imshow(avg)
```

```python
gray_im2 = np.average(image, axis=2)
plt.imshow(gray_im2, cmap='gray')
plt.show()
```

```python
from skimage import color
internal_gray = color.rgb2gray(image)
io.imshow(internal_gray)
```
