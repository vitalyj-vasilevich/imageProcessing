```python
from skimage import data
from skimage import color
from skimage.util import pad
from skimage.transform import resize
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

image = io.imread('Images\\CursedCat.JPG')
io.imshow(image)
```

```python
def my_median_filter(img, x_sz, y_sz):
    result_img = img.copy()
    for i in range(img.shape[0] - y_sz + 1):
        for j in range(img.shape[1] - x_sz + 1):
            raw_cells = np.array([])
            raw_cells = np.append(raw_cells, img[i:i + y_sz,j:j + x_sz])        
            raw_cells.sort()
            result_img[i][j] = cells_median(raw_cells)
    result_img = result_img[:img.shape[0] - y_sz, :img.shape[1] - x_sz]
    return result_img

def cells_median(cells):
    middle_index = cells.shape[0] // 2
    if(cells.shape[0] % 2 == 0):
        return (cells[middle_index - 1] + cells[middle_index + 1]) / 2
    else:
        return cells[middle_index // 2 + 1]
```

```python
median_filtered_image = my_median_filter(image, 5, 5)
io.imshow(median_filtered_image)
```
