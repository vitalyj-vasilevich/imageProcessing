```python
from skimage import io
from skimage import color
from skimage import transform
import numpy as np

image = color.rgb2gray(io.imread('Images\\cat1000.jpg'))
io.imshow(image)
```

```python
def prep_one_parameter(image, param):
    result_img = image.copy()
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if(image[i][j] < param):
                image[i][j] = 0
    return result_img
```

```python
one_param_image = prep_one_parameter(image, 0.5)
io.imshow(one_param_image)
```

```python
def prep_two_parameters(image, param_1, param_2):
    result_img = image.copy()
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            if(image[i][j] < param_1):       
                image[i][j] = 0            
            if(image[i][j] > param_2):
                image[i][j] = 1
                
    return result_img
```

```python
two_params_image = prep_two_parameters(image, 0.4, 0.6)
io.imshow(two_params_image)
```

```python
def table_method(image, t):
    x = np.arange(256)
    table = np.where(x > t, x, 0)
    return table[image]
```

```python
image = (255 * image).astype(np.uint8)
io.imshow(table_method(image, 1))
```
